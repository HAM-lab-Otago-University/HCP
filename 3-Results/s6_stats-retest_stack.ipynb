```python
from IPython.display import display
import numpy as np
import pandas as pd
import os
import pickle
```

```python
path_input = '/home/narunlab/Documents/Alina/jj/'

with open(path_input+'retest_layer2_output.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
for fir_reg in ['svr','rf','xgb', 'eNet']:
    for sec_reg in ['svr','rf','xgb', 'eNet']:
        print('\n\n##################fir_reg:',fir_reg,'     sec_reg:',sec_reg,'##################')
        for show_set in ['test1','test2']:
            print(show_set)
            stat = pd.DataFrame( columns=['r2', 'mse'])
            for smod in ['all', 'task', 'non-task', 'top-task']:
                stat.loc[smod] = [data[fir_reg][sec_reg][smod]['r2_'+show_set],
                                  data[fir_reg][sec_reg][smod]['mse_'+show_set]
                                                   ]
            display(stat)
```

# To get the predictions

The data is stored in a 4 layer dictionary: 
- fir_reg: the algorithm used on the first layer
- sec_reg: the algorithm used on the second layer
- smod: the stacked-modalities you want to see

you can get the predicted value directly using:
- data[fir_reg][sec_reg][smod]['pred_test1']
- data[fir_reg][sec_reg][smod]['pred_test2']
